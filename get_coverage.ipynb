```python
from datasets import TTSDataset
import json

dataset = TTSDataset('gsm8k', 0.5)
cal_set = dataset.cal_dataset
pred_set = dataset.test_dataset
```

```python
pred_questions = [data['question'] for data in cal_set]
```

```python
with open(f"outputs_exp/s1.1-3B_gsm8k_min_tokens_fixed.jsonl", "r") as f:
    lines = f.readlines()
    for l in lines:
        line = json.loads(l)
        q = line['question']
        if q in pred_questions:
            if line['flag'] == "Wrong":
                pred_questions.remove(q)
```

```python
counter = 0
with open(f"result/Calibrate_s1.1-3B_gsm8k_final.jsonl", "r") as f:
    lines = f.readlines()
    for l in lines:
        line = json.loads(l)
        q = line['question']
        if q in pred_questions:
            if line['flag'] == "Correct":
                counter += 1
            else:
                print(line["ground_truth"], line["final_answer"])
counter / len(pred_questions)
```
